Question 1 : What is Power BI and why is it used in businesses?

- Power BI is a business intelligence tool developed by Microsoft that helps users analyze data and create interactive reports and dashboards. It allows users to connect to different data sources, transform the data, and visualize it in an easy-to-understand format.

Businesses use Power BI because it helps them make better decisions based on data. It provides real-time insights, improves data analysis, and helps in tracking business performance through visual reports.

Question 2 : Name and explain the three main components of Power BI.

- Power BI Desktop
It is used to create reports and dashboards on a computer. It allows users to import data, clean it, and design visualizations.
Power BI Service
It is an online platform where reports can be published and shared with others. It helps in collaboration and accessing reports from anywhere.
Power BI Mobile
It allows users to view reports and dashboards on mobile devices like phones and tablets.

Question 3 : Explain the Power BI workflow.

- Data Collection – Import data from different sources like Excel, databases, or APIs.
Data Transformation – Clean and prepare data using Power Query.
Data Modeling – Create relationships between tables.
Visualization – Build charts, graphs, and dashboards.
Publishing – Upload reports to Power BI Service.
Sharing – Share reports with team members.

Question 4 : List any four data cleaning tasks that can be performed in Power Query.

- Removing duplicate records

- Handling missing values

- Changing data types

- Splitting or merging columns

These tasks help in preparing clean and structured data.

Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.

-  Data View
Used to see the data in table format. It helps in checking and editing data.
- Report View
Used to create visualizations like charts, graphs, and dashboards.
- Model View
Used to manage relationships between different tables in the dataset.

Each view plays an important role in building reports.
