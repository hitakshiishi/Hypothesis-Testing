Q1. Create a New Database and  Table for Employees.

```python
import sqlite3

```

```python
conn = sqlite3.connect('company.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("DROP TABLE IF EXISTS employees")
```

```python
cursor.execute("""
CREATE TABLE employees (
    employee_id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    department TEXT,
    salary INTEGER,
    hire_date TEXT
)
""")
```

```python
conn.commit()
```

```python
conn.close()
```

```python
print("Database and table created successfully!")
```

Q3. Display All Employee Records Sorted by Salary (Lowest to Highest).
- he ORDER BY clause is used to sort records in ascending order based on the salary column.
ASC means ascending order (lowest to highest).
Therefore, employees are displayed starting from the lowest salary to the highest salary.

```python
import sqlite3

conn = sqlite3.connect('company.db')
cursor = conn.cursor()
```

Q4. Show Employees Sorted by Department (A–Z) and Salary (High → Low)

Q6. Create and Populate a Sales Table

```python
import sqlite3

conn = sqlite3.connect('company.db')
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    sale_id INTEGER PRIMARY KEY,
    customer_name TEXT,
    amount INTEGER,
    sale_date TEXT
)
""")
```

```python
cursor.execute("DELETE FROM sales")  # Optional: clear old data

sales_data = [
(1, 'Aditi', 1500, '2024-08-01'),
(2, 'Rohan', 2200, '2024-08-03'),
(3, 'Aditi', 3500, '2024-09-05'),
(4, 'Meena', 2700, '2024-09-15'),
(5, 'Rohan', 4500, '2024-09-25')
]

cursor.executemany("INSERT INTO sales VALUES (?, ?, ?, ?)", sales_data)

conn.commit()
print("Sales table created and populated successfully.")
```

Q9. What is the Difference Between a Primary Key and a Foreign Key?

- Primary Key

  A Primary Key is a column (or a set of columns) that uniquely identifies each record in a table.

- Key Characteristics:

   1.Must contain unique values

   2.Cannot contain NULL values

   3.Only one primary key is allowed per table

   4.Ensures entity integrity.

   5.only one per table.

- Foreign Key

  A Foreign Key is a column that creates a relationship between two tables. It refers to the Primary Key of another table.

- Key Characteristics:

  1.Can contain duplicate values

  2.Can contain NULL values (unless restricted)

  3.Maintains referential integrity

  4.Links related tables together.

  5.Multiple allowed

Q10. What Are Constraints in SQL and Why Are They Used?
- In SQL, constraints are rules applied to table columns to control the type of data that can be inserted into a database. They are used to ensure data accuracy, consistency, and integrity within a table.

  Constraints act as restrictions that prevent invalid or unwanted data from being stored in the database.
- Why Are Constraints Important?

Constraints are important because they:

Maintain data integrity

Prevent incorrect or duplicate entries

Ensure consistency across tables

Enforce business rules

Improve database reliability

Protect relationships between tables

Without constraints, databases may contain incorrect or meaningless data.
Types of Constraints in SQL.

- 1️ PRIMARY KEY

A Primary Key uniquely identifies each record in a table.

Must contain unique values

Cannot contain NULL values

Only one primary key is allowed per table .

- 2 FOREIGN KEY

A Foreign Key is used to create a relationship between two tables.

References the Primary Key of another table

Maintains referential integrity

Prevents invalid data insertion
- 3 NOT NULL

The NOT NULL constraint ensures that a column cannot store NULL values.
- 4 CHECK

The CHECK constraint applies a condition to restrict the values that can be entered.
