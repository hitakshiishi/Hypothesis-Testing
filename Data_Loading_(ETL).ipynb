Q7. Pre-Load Validation Checklist

Before loading the dataset, the following checks should be performed:

Verify primary key uniqueness

Detect and remove duplicate records

Check for missing values

Validate numeric fields

Standardize date formats

Validate data types

Check referential integrity

Perform data quality checks

These validations ensure the dataset is safe to load.

Q8. Cleaning Strategy

Steps to make the dataset load-ready:

Remove duplicate record with Order_ID O101

Replace Null Sales_Amount with estimated value or remove record

Convert “Three Thousand” → 3000

Standardize date format to YYYY-MM-DD

Validate numeric columns

Verify primary key uniqueness

Re-run validation checks

After cleaning, the dataset can be safely loaded.

Q9. Loading Strategy Selection
a) Which loading strategy should be used?

Incremental Load.

b) Justification

Since the dataset represents daily sales data, incremental loading is better because:

Only new records are added

Processing time is reduced

Historical data remains unchanged

ETL pipelines become more efficient
