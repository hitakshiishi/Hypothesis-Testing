Q1. What are the most common reasons for missing data in ETL pipelines?



- Missing data can occur in ETL pipelines for several reasons. One common reason is data entry errors, where users forget to enter values or leave fields empty. Another reason is system integration issues, where data coming from different systems may not match properly. Missing data can also occur due to technical failures, such as interrupted data transfers or system crashes. Sometimes privacy or business rules also prevent certain information from being stored, which leads to missing values in the dataset.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

Blindly deleting rows with missing values is considered a bad practice because it can lead to loss of important data. If many rows are deleted, the dataset may become too small or biased. This can affect the accuracy of analysis and lead to incorrect conclusions. Instead of deleting rows without thinking, it is better to analyze why the data is missing and apply appropriate techniques such as imputation or data correction.

Q3. Explain the difference between:

Listwise deletion
column wisw deletion

Also mention one scenario where each is appropriate.

- Listwise deletion means removing entire rows from a dataset if they contain missing values. This method is useful when only a small number of rows have missing data and removing them does not affect the dataset significantly.

- Column deletion, on the other hand, means removing an entire column if it contains too many missing values and is not very useful for analysis.

- Example scenario:
Listwise deletion is suitable when only a few records are incomplete.
Column deletion is suitable when a column has a large number of missing values and cannot provide meaningful insights.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

- Median imputation is preferred for skewed data because the mean can be heavily affected by extreme values or outliers. In datasets such as income, some individuals may earn extremely high salaries which can distort the mean. The median represents the middle value of the dataset and is less affected by outliers. Therefore, using the median provides a more realistic estimate for missing values in skewed datasets.

Q5. What is forward fill and in what type of dataset is it most useful?

- Forward fill is a technique used to handle missing values by filling them with the previous available value in the dataset. This method is commonly used in time-series data, such as stock prices, sales records, or sensor data, where the previous value is likely to remain valid until a new value appears.

- for example, if sales data for a particular day is missing, it can be filled with the sales value from the previous day.

Q6. Why should flagging missing values be done before imputation in an ETL workflow?

- Flagging missing values before imputation is important because it helps track which data was originally missing. When missing values are replaced with estimated values, the information about missing data can be lost. By creating a flag column, analysts can later identify which records were imputed and analyze patterns in missing data. This improves transparency and helps maintain data quality.

Q7. Consider a scenario where income is missing for many customers. How can this missingness itself provide business insights?

- Missing income values can provide useful insights about customer behavior. For example, customers may choose not to share income information due to privacy concerns or because they belong to certain demographic groups. If income is missing for a particular region or customer segment, it may indicate issues in data collection or customer trust. Businesses can analyze these patterns to improve data collection strategies or understand customer characteristics better.
