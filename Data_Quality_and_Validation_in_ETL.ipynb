Question 1.Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

- Data quality in ETL pipelines refers to the accuracy, completeness, consistency, and reliability of the data being processed and loaded into a data warehouse. High-quality data ensures that the information used for analysis and reporting is correct and trustworthy.

- Data quality is more than just data cleaning because it involves several processes such as validation, standardization, monitoring, and error detection. Cleaning only removes errors or duplicates, but maintaining data quality also ensures that the data follows business rules, maintains integrity, and remains consistent across systems.

Question 2.Explain why poor data quality leads to misleading dashboards and incorrect decisions.

- poor data quality can significantly affect business decisions. Dashboards and reports depend on accurate data to display insights. If the data contains errors, missing values, or duplicates, the visualizations created from it will also be incorrect.

- For example, if sales data contains duplicate transactions, the total revenue shown on a dashboard will be higher than the actual value. This may lead managers to believe that the business is performing better than it actually is. As a result, decisions based on such incorrect insights can negatively affect business strategy and planning.

Question 3.What is duplicate data? Explain three causes in ETL pipelines.

- Duplicate data refers to records that appear more than once in a dataset even though they represent the same real-world event or entity.

- Three common causes of duplicate data in ETL pipelines are:

- Multiple data sources – When data is collected from different systems, the same record may appear in more than one source.

- Improper joins or transformations – Incorrect ETL transformations may accidentally create repeated rows.

- Repeated data loads – If an ETL job runs multiple times without proper checks, the same records may be loaded again into the system.

- Duplicate data can lead to incorrect analysis and should be identified and removed.

Question 4.Differentiate between exact, partial, and fuzzy duplicates.

- Exact duplicates occur when two records are completely identical in every column. For example, two rows having the same customer ID, product ID, transaction amount, and date.

- Partial duplicates occur when some fields are the same but others differ slightly. For example, the same customer purchasing the same product on the same day but with a slightly different quantity.

- Fuzzy duplicates occur when the records represent the same entity but contain small differences such as spelling mistakes or formatting changes. For example, “Rahul Mehta” and “Rahul M.” referring to the same person.

Question 5.Why should data validation be performed during transformation rather than after loading?

- Data validation should be performed during the transformation stage because errors can be detected and corrected before the data enters the final system. If validation happens only after loading, incorrect data may already be stored in the database and affect reports or analytics.

- Performing validation during transformation helps ensure that only clean and accurate data moves forward in the ETL pipeline. This reduces the risk of data corruption and improves overall data reliability.

Question 6.Explain how business rules help in validating data accuracy. Give an example.

- Business rules are predefined conditions that ensure the data follows certain logical requirements before it is accepted into the system. These rules help verify whether the data values are valid and meaningful.

- For example, a business rule might state that the transaction amount must always be greater than zero. If a record contains a negative transaction amount, the ETL process will identify it as invalid and either correct it or reject the record.

- Business rules help maintain consistency and accuracy across datasets.

Question 7

SQL query to list duplicate keys using the business key

```python
import pandas as pd

data = {
    "Customer_ID":["C101","C102","C101","C103","C104","C105","C106","C101"],
    "Product_ID":["P11","P12","P11","P13","P14","P15","P16","P11"],
    "Txn_Amount":[4000,1500,4000,6000,2500,None,1800,4000],
    "Txn_Date":["2025-12-01","2025-12-01","2025-12-01","2025-12-02","2025-12-02","2025-12-03",None,"2025-12-01"]
}

df = pd.DataFrame(data)

duplicates = df.groupby(
["Customer_ID","Product_ID","Txn_Date","Txn_Amount"]
).size().reset_index(name="duplicate_count")

duplicates = duplicates[duplicates["duplicate_count"] > 1]

duplicates
```
