Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?

- Data loading in Power BI means importing data from different sources like Excel, CSV, or databases into Power BI.

It is considered the first step because without loading data, we cannot perform analysis, create reports, or build dashboards. All further steps like cleaning, transformation, and visualization depend on this.

Question 2 : Explain the difference between “Load” and “Transform Data” in Power BI.

-  Load means directly importing data into Power BI without making any changes.

Transform Data means opening Power Query Editor to clean and modify data before loading.

In simple words:

Load → Direct import
Transform → Clean and prepare data first

Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset.

- A Fact Table contains numerical values (measures).

Example:

Confirmed cases

Deaths

Recovered

A Dimension Table contains descriptive information.

Example:

Country

State

Date

Fact table answers “how much”, dimension table answers “where/when”.

Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?

- Star schema is preferred because:

Simple structure

Faster performance

Easy to understand

Better for reporting

Snowflake schema is complex and slows down queries.

Question 6 : Identify and replace null values in Vaccination_Status.

- Steps:
Select column with null values

Go to:

Transform → Replace Values

Replace:

null → 0

For text columns (like Vaccination_Status):

null → "Unknown"
