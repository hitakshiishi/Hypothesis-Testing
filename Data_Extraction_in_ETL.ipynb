Question 1 : Describe different types of data sources used in ETL with suitable examples.

In ETL (Extract, Transform, Load), data is collected from different sources before processing and loading into a data warehouse.

 Data Sources:

1. Databases

Structured data stored in relational databases.

Examples: MySQL, Oracle.

Example: Extracting customer records from a banking database.

2. Files

Data stored in file formats.

Examples: CSV, Excel, JSON

Example: Importing sales data from a CSV file.

3. APIs

Applications provide data through APIs.

Used for real-time data extraction.

Example: Extracting weather data using a weather API.

4. Cloud Storage

Data stored in cloud platforms.

Examples:

Google Cloud Storage

AWS

Azure Blob Storage

5. Web Data

Data extracted from websites using web scraping.

Example: Collecting product prices from e-commerce websites.

Question 2 : What is data extraction? Explain its role in the ETL pipeline.

- Data extraction is the process of collecting data from different sources such as databases, files, APIs, or applications.

- Role in ETL Pipeline:

Extract – Data is collected from multiple sources.

Transform – Data is cleaned and converted into a suitable format.

Load – Data is stored into a data warehouse or database.

- Importance of Data Extraction:

Collects raw data from different systems.

Enables data integration.

Ensures data is available for analysis and reporting.

Question 5 : Explain three common challenges faced during data extraction.

1. Data Quality Issues

Data may contain missing values, duplicates, or incorrect formats.

Example: Customer names entered differently.

2. Large Data Volumes

Extracting very large datasets can slow down processing.

Example: Extracting millions of transaction records.

3. Data Source Compatibility

Different systems may use different formats.

Example: Combining data from CSV files and databases.

Question 6 : What are APIs? Explain how APIs help in real-time data extraction.

An API (Application Programming Interface) allows different applications to communicate and exchange data.

How APIs Help in Real-Time Extraction:

Provide live data updates.

Allow automated data retrieval.

Support integration between systems.

Example:
A stock market API provides real-time stock prices.

Question 7 : Why are databases preferred for enterprise-level data extraction?

2. Fast Query Processing

3. Data Security

4. Scalability

5. Data Integrity

Ensures data accuracy and consistency.

Question 8 : What steps should an ETL developer take when extracting data from large CSV files (1GB+)?


1. Use Streaming or Chunk Processing

Read data in small batches instead of loading the whole file.


2. Remove Unnecessary Columns

Extract only required data.

3. Use Efficient Tools

Use tools like:



Hadoop

Python Pandas

4. Validate Data During Extraction

Check for missing or corrupted records.

5. Compress Files if Needed
