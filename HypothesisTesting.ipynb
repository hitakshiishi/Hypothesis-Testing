Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?
 - The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship between variables. It represents the default assumption that any observed change is due to random chance.
- Importance:

  1.It provides a starting point for statistical testing.

  2.Helps researchers make objective decisions using data.

  3.Prevents drawing conclusions based on assumptions.

  4.Acts as a benchmark to determine whether results are statistically significant.



Question 2: What does the significance level (α) represent in hypothesis testing?
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

 It represents the risk of making a wrong decision.

- Common Value:

  α = 0.05 (5%)

  This means we accept a 5% chance of concluding that a result exists when it actually does not.

- Why it matters:

  1.Helps control errors in decision-making.

  2.Makes hypothesis testing more reliable.

  3.Lower α → More strict testing.

Question 3: Differentiate between Type I and Type II errors.
-Type I error is usually considered more serious because it claims something false is true.
- Type I Error
1.Rejecting a true null hypothesis.
2.Also called False Positive.
3.Probability = α
4.Example: Saying a medicine works when it actually doesn’t.
- Type II Error
1.Accepting a false null hypothesis.
2.Also called False negative.
3.Probability = β
4.Saying a medicine doesn’t work when it actually does.

Question 4: Explain the difference between a one-tailed and two-tailed test.
- one-tailed test:
A one-tailed test checks for a difference in one specific direction (greater OR less).

  Example:
  H₀: Average salary ≤ 50,000
  H₁: Average salary > 50,000

  Used when we are only interested in one outcome.
- two-tailed test:
A two-tailed test checks for a difference in both directions.

  Example:
  H₀: Average weight = 70 kg
  H₁: Average weight ≠ 70 kg

  Used when any change (increase or decrease) matters.


Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.
- Since the sample size is small, we use a t-test.

- Step 1: Hypotheses
H₀: μ = 10
H₁: μ ≠ 10

(Two-tailed test)

- Step 2: Calculate Test Statistic
- Step 3: Critical Value

  Degrees of freedom = n−1 = 8

  From t-table at α = 0.05 → ±2.306
- Step 4: Decision
  Since 2 < 2.306, we fail to reject H₀.


```python
import scipy.stats as stats

sample_mean = 12
pop_mean = 10
std_dev = 3
n = 9

t_stat = (sample_mean - pop_mean) / (std_dev / (n**0.5))
p_value = stats.t.sf(abs(t_stat), df=8)*2

t_stat, p_value
```

Question 6: When should you use a Z-test instead of a t-test?
- A Z-test is used when:

1.Sample size is greater than 30

2.Population standard deviation is known

3.Data follows a normal distribution
- A t-test is used when:

1.Sample size is small (<30)

2.Population standard deviation is unknown

In real-world problems, t-tests are more common because population deviation is rarely known.

Question 7:The productivity of 6 employees was measured before and after a training program.
-This is a paired t-test because the same employees are measured before and after training.

Data:

Before: 50, 60, 58, 55, 62, 56

After: 55, 65, 59, 58, 63, 59

- Step 1: Hypotheses

H₀: Training did NOT improve productivity.

H₁: Training improved productivity.

- Step 2: Differences

5, 5, 1, 3, 1, 3

Mean difference = 3

t-value ≈ 4.24

Critical value (df=5, α=0.05) ≈ 2.015

- Step 3: Decision

Since 4.24 > 2.015, reject H₀.


The training program significantly improved employee productivity.

```python
import scipy.stats as stats

before = [50,60,58,55,62,56]
after = [55,65,59,58,63,59]

stats.ttest_rel(after, before)
```

Question 8: A company wants to test if product preference is independent of gender.
- Step 1: Hypotheses

H₀: Product preference is independent of gender.

H₁: Product preference depends on gender.

- Step 2: Chi-Square Test

Calculated χ² ≈ 16.67

Critical value at df=1, α=0.05 → 3.84

- Step 3: Decision

Since 16.67 > 3.84, reject H₀.

There is a significant relationship between gender and product preference.

Gender influences which product customers choose.

```python
import scipy.stats as stats

data = [[30,20],
        [10,40]]

chi2, p, dof, expected = stats.chi2_contingency(data)

chi2, p
```
