Q 4.Explain the difference between Sorting and Filtering with an example from this dataset.

- Sorting:
Arranges data (e.g., highest sales to lowest).

- Example:
Sort Sales column descending.

- Filtering:
Shows only specific data.

- Example:
Filter Region = East only.

- Difference:
Sorting rearranges data. Filtering hides unwanted data.
