Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.
- Covariance: is a statistical measure that indicates the direction of the relationship between two variables. It shows whether the variables increase together or if one increases while the other decreases.

  1.Indicates only direction.

  2.Value ranges from −∞ to +∞.

  3.Difficult to interpret due to scale.

  4.Depends on units
- Correlation: on the other hand, measures both the strength and direction of the relationship between variables.

  1.Indicates both strength and direction.

  2.Value always lies between −1 and +1.

  3.Easy to interpret.

  4.Unit-free measure.

Therefore, correlation is often preferred because it provides clearer insight into how strongly variables are related.

Q2. What does a positive, negative, and zero covariance indicate?
- Positive Covariance: Both variables move in the same direction.
Example: Higher education → Higher income.

- Negative Covariance: Variables move in opposite directions.
Example: Higher product price → Lower demand.

- Zero Covariance: No relationship exists between the variables.

 Covariance mainly helps identify whether a relationship exists, but not how strong it is.

Q3. Discuss the limitations of covariance. Why is correlation preferred?
- Limitations of Covariance:

  1.Values are not standardized.

  2.Hard to interpret magnitude.

  3.Depends on measurement units.

  4.Cannot determine strength clearly.

- Why Correlation is Preferred:

  1.It is standardized between −1 and +1.

  2.Easy to compare relationships.

  3.Provides both strength and direction.

  4.Independent of units.

Because of these advantages, correlation is widely used in data analysis and research.

Q4. Difference between Pearson’s and Spearman’s correlation coefficients.
- Pearson’s Correlation:

  1.Measures linear relationships.

  2.Uses actual numerical values.

  3.Best for normally distributed data.

- Spearman’s Rank Correlation:

  1.Uses ranked data instead of raw values.

  2.Measures monotonic relationships.

  3.Works well with outliers or non-normal data.

- When to Use Spearman:

  1.When data is ordinal
  
  2.When outliers are present
  
  3.When the relationship is not linear

 Spearman is more flexible when assumptions for Pearson are not met.

Q5. If correlation between X and Y is 0.85, interpret it. Can we infer causation?

A correlation coefficient of 0.85 indicates a strong positive relationship between X and Y. This means that as X increases, Y is also likely to increase.

However, correlation does NOT imply causation.

Just because two variables move together does not mean one causes the other. External factors may influence both.

Example: Ice cream sales and sunglasses sales rise together — but one does not cause the other; hot weather affects both.
