Question 1 :

What is a Calculated Field and why is it important in business dashboards?

A calculated field in Tableau is a new field created using formulas based on existing data. It is important because it allows users to perform custom calculations such as profit, ratios, or classifications, which help in better analysis and decision-making.

Question 2 :

 Why can’t parameters change visuals directly in Tableau?

Parameters cannot directly change visuals because they are independent variables. They need to be connected with calculated fields or filters to affect the visualization.

Question 3 :

When should EXCLUDE LOD be used?

EXCLUDE LOD is used when we want to remove a dimension from the calculation level while still keeping it in the view. It helps in comparing aggregated values at a higher level.

Question 4 :

Difference between FIXED and INCLUDE LOD expressions?

FIXED → Ignores filters and calculates at a fixed level
INCLUDE → Adds extra detail to calculation

FIXED is independent, while INCLUDE depends on view level.
