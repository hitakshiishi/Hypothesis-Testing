Q8. What is the difference between a Worksheet and a Dashboard in Tableau?

Why is a dashboard more useful for decision-makers?

- A Worksheet in Tableau is used to create a single visualization such as a bar chart, line chart, or map. It represents one view of data.

A Dashboard is a collection of multiple worksheets displayed together on one screen. It combines different charts and visuals to provide a complete overview of data.

| Worksheet                    | Dashboard                |
| ---------------------------- | ------------------------ |
| Single chart                 | Multiple charts          |
| Shows one analysis           | Shows complete analysis  |
| Used for basic visualization | Used for decision-making |

- A dashboard is more useful for decision-makers because it shows all important information in one place. It allows users to compare data, identify trends, and make quick decisions without switching between multiple sheets.

- Dashboards also support interaction, such as filters and highlights, which helps in deeper analysis.

Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)

- Tableau Prep is a data preparation tool used to clean, transform, and organize raw data before analysis. It helps in improving data quality by removing errors, handling missing values, and making the data ready for visualization.

- It is important in real-world data projects because raw data is often messy and inconsistent. Tableau Prep ensures that the data is accurate, clean, and structured, which leads to better analysis and reliable results.

1. Remove Null Values

To remove null values, the dataset is filtered to exclude rows where the selected column contains null values. This ensures that incomplete data does not affect analysis.

2. Remove Duplicate Records

Duplicate records can be removed by identifying repeated rows and eliminating them from the dataset. This ensures that data is not counted multiple times and maintains accuracy.

3. Change Incorrect Data Types

Incorrect data types can be corrected by converting fields into appropriate formats such as converting text to number or date format. This ensures that calculations and analysis are performed correctly.
