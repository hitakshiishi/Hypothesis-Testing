Question 1 : Define Data Transformation in ETL and explain why it is important.

Data Transformation is the process of converting raw data into a clean, structured, and useful format before loading it into a data warehouse or database. It is the second stage of the ETL process (Extract, Transform, Load).

During transformation, data is cleaned, filtered, organized, and standardized so that it can be easily analyzed.

Data transformation is important because raw data often comes from multiple sources and may contain errors, duplicates, or different formats. Transformation helps to improve data quality, maintain consistency, and make the data suitable for reporting and analysis. Without transformation, the extracted data may not be reliable for business decision-making.

Question 2 : List any four common activities involved in Data Cleaning.

- Removing duplicate records – Eliminating repeated data entries to maintain accuracy.

- Handling missing values – Filling missing values using methods like mean, median, or removing the rows.

- Correcting inconsistent formats – Standardizing formats such as date formats, gender values, or text capitalization.

- Removing irrelevant data – Deleting unnecessary columns or records that are not useful for analysis.

- These activities help ensure that the dataset is accurate, consistent, and reliable

Question 3 : What is the difference between Normalization and Standardization?

Normalization and Standardization are techniques used to scale numerical data.

Normalization scales the data values to a fixed range, usually between 0 and 1. It is useful when we want to compare values on the same scale.

Question 4 : A dataset has missing values in the “Age” column. Suggest two techniques to handle this and
explain when they should be used.

There are several techniques to handle missing values in the Age column.

One method is mean or median imputation, where the missing age values are replaced with the average or median age of the dataset. This method is useful when the dataset is large and the missing values are few.

Another method is removing rows with missing values. This technique is useful when the number of missing values is very small and removing them will not significantly affect the dataset.

Choosing the correct method depends on the size of the dataset and the importance of the missing information.

- In this technique, each category is converted into a separate column with binary values (0 or 1).

Question 7: Explain the difference between Data Integration and Data Mapping in ETL.

- Data Integration is the process of combining data from multiple sources into a single unified dataset. It ensures that data from different systems works together.

- Example: Combining customer data from sales, marketing, and finance databases.

- Data Mapping is the process of linking fields from one data source to another. It defines how data from the source system will correspond to the destination system.

- Example: Mapping “Cust_ID” from a source database to “Customer_ID” in a data warehouse.

- data integration merges data sources, while data mapping defines relationships between fields.

Question 8: Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.

- Z-score Standardization is often preferred when datasets contain outliers because it considers the mean and standard deviation of the data.

- Min-Max Scaling compresses all values into a fixed range between 0 and 1. If there are extreme outliers, the majority of data points get squeezed into a very small range.

- Z-score Standardization handles outliers better because it scales data based on how far each value is from the mean. This allows the model to maintain meaningful differences between data points even when extreme values are present.

Therefore, Z-score Standardization is more suitable when the dataset contains outliers.
