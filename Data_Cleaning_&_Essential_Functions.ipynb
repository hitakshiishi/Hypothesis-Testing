Q1.What is data cleaning, and why is it important in data analysis?

What are the potential consequences of analyzing unclean or messy data?

Explain the common steps involved in cleaning and organizing data.


-What is Data Cleaning?

  Data cleaning is the process of identifying and correcting errors, inconsistencies, duplicates, and missing values in a dataset to ensure it is accurate and reliable for analysis.

- Why is it Important?

  Improves data accuracy

  Removes duplicate entries

  Fixes formatting issues

  Prevents wrong analysis results

  Saves time in reporting

- Potential Consequences of Analyzing Unclean Data

  Incorrect business decisions

  Misleading insights

  Duplicate calculations

  Reduced data credibility

  Financial losses

- Common Steps in Data Cleaning

  Remove duplicate values

  Handle missing data

  Standardize formatting

  Correct spelling errors

  Trim extra spaces

  Validate data types

  Apply filters and sorting

TODAY()
Returns current date


NOW()
Returns date + time

Example:

Use TODAY() for deadlines

Use NOW() for timestamps

b) Days Left Until 2025-06-10

If A1 = 2025-06-10:
