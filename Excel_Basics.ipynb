1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.
- 1. Relative Cell Reference (A1)

  A relative reference changes automatically when the formula is copied to another cell.

- 2. Absolute Cell Reference ($A$1)

  An absolute reference does NOT change when copied.

- 3. Mixed Cell Reference (A$1 or $A1)

  One part is fixed, the other changes.
